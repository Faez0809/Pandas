# #append() funcion


In Pandas 2.0 and above, the .append() method was removed.

now you must use pd.concat() instead.




```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Maths':[45,78,45,90,66], 'Physics':[33,67,12,90,44]})
df2 = pd.DataFrame({'Roll No.':[6,7,8,9,10], 'Maths':[78,73,45,90,69], 'Physics':[23,67,88,0,98]})
print(df1)
print(df2)

```

```python
df_new = pd.concat([df1,df2])
df_new
```

```python
df_new = pd.concat([df1,df2],ignore_index=True)
df_new
```

```python
df_new = pd.concat([df1,df2],ignore_index = True, sort =True)
df_new
```

```python
df3 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Chemistry':[45,78,45,90,66], 'Physics':[33,67,12,90,44]})
df4 = pd.DataFrame({'Roll No.':[6,7,8,9,10], 'Maths':[78,73,45,90,69], 'Physics':[23,67,88,0,98]})
print(df3)
print(df4)
```

```python
df_new = pd.concat([df3,df4],ignore_index = True)
df_new
```
