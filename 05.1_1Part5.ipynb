#Filling the Null Values

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/sample.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({'physics':'none','chemistry':0,'Maths':30})
```

```python
#Forward fill ( filled with prev row value) as by defaule axis=0

df.fillna(method = 'ffill')
```

```python
df.fillna(method = 'ffill', axis = 1) # it will fill from prev column
```

```python
#we can fill with the avg value

df['Physics'].fillna(value=df['Physics'].mean())
```

```python
#bfill will fill the value with the next value

df.fillna(method = 'bfill')
```

```python
#using inplace = True then it will actually modify my original dataframe

df.fillna(method = 'bfill', inplace = True)
df
```
