# #Pandas DataFrame

```python
import pandas as pd
```

Creating an empty DF

```python
df = pd.DataFrame()
print(df)
```

#DataFrame Using List

```python
lst = [1,2,3,4,5]
df = pd.DataFrame(lst)
print(df)
```

DataFrame with multiple columns
provided with list of two lists
---



```python
lst = [[1,2,3,4,5],[11,12,13,14,15]]
df = pd.DataFrame(lst)
df
```

With Dictionaries:

taken a list of two dictionaries

Dictionary keys represents Column names

```python
a = [ {'a':5,'b':7,'c':9,'d':2},
      {'a':4,'b':8,'c':19,'d':12} ]

df=pd.DataFrame(a)
df
```

Creating a DataFrame using Pandas Series

Every series will be treated as a column

```python
b = { 'RollNo.' :pd.Series([1,2,3,4,5]),
       'Maths'  :pd.Series([67,89,23,90,56]),
       'Physics':pd.Series([12,98,44,90,78])

  }

df = pd.DataFrame(b)
df
```
