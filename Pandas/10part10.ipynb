# #Pivot Table

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sample2.csv')
df.head()
```

```python
df1=df[['Roll No.','Branch','Physics','Chemistry','Maths','Computer']]
df1
```

```python
pd.pivot_table(df1, index = 'Branch')
```

```python
pd.pivot_table(df1, index ='Branch', aggfunc='sum')
```

```python
pd.pivot_table(df, index = 'Branch', aggfunc='count')
```

```python
pd.pivot_table(df, index = 'Branch',aggfunc='max')
```

```python
pd.pivot_table(df1, index ='Branch', columns='Roll No.', aggfunc='mean')
```
